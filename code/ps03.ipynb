2) Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?

```python
def make_wave(offset):
    signal = thinkdsp.SinSignal(freq=440, offset=offset)
    wave = signal.make_wave(duration=0.5, framerate=10000)
    return wave

def compute_corr(offset):
    wave1 = make_wave(0)
    wave2 = make_wave(offset=-offset)
    
    thinkplot.preplot(2)
    wave1.segment(duration=0.01).plot()
    wave2.segment(duration=0.01).plot()
    
    corr = wave1.corr(wave2)
    print('corr =', corr)
    
    thinkplot.config(xlabel='time (s)', legend=False, ylim=[-1.05, 1.05])

def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

PI2 = 2 * math.pi
offsets = numpy.linspace(0, PI2, 101)

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```

```python
lags, corrs = autocorr(coin_wave)
thinkplot.plot(lags, corrs)
```

The autocorrelation function does not decay exponentially, for the first segment (from lag of 0 to ~400) appears to be about linear, with peaks at lag = 450 days of 0.5 for no correlation, and a small peak of less anti-correlation at 700 days.  So I'd say on the daily prices there is not evidence of periodic behavior.

4) Test the algorithmic complexity of analyze1, analyze2 and scipy.fftpack.dct.

```python
import scipy.fftpack
import dct

signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)
ns = 2 ** numpy.arange(4, 12)
print(ns)
results = []
for N in ns: #for each of the 2^n in range
    ts = (0.5 + numpy.arange(N)) / N 
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o dct.analyze1(ys, freqs, ts) #decompose a noise signal evaluated at 2^n points
    results.append(result) #append time to compute 

bests = [result.best for result in results]
bests
```

```python
thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log')
x = numpy.log(ns)
x = ns
y = numpy.log(bests)
t = scipy.stats.linregress(x,y)
t
```

```python
thinkplot.plot(ns,bests)
```

So the time complexity is about O(n^2), from the first arg of t.  The function dct.analyze1 has quadratic growth, given these parameters.  But analyze1 uses linsolve, which should be a cubic operation.  What if we look at a greater range?

Can we be sure this is quadratic?

```python
import scipy.fftpack
import dct

signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)
ns = 2 ** numpy.arange(4, 14) #to 14 takes about 4 minutes in total
print(ns)
results = []
for N in ns: #for each of the 2^n in range
    ts = (0.5 + numpy.arange(N)) / N 
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o dct.analyze1(ys, freqs, ts) #decompose a noise signal evaluated at 2^n points
    results.append(result) #append time to compute 

bests = [result.best for result in results]
bests
```
