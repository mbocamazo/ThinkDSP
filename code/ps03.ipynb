Compute the power spectrum of BitCoin prices. Does it resemble white, pink, or Brownian noise?

```python
nrows = 1625
df = pandas.read_csv('coindesk-bpi-USD-close.csv', nrows=nrows, parse_dates=[0])
ys = df.Close.values
coin_wave = thinkdsp.Wave(ys, framerate=1)
coin_wave.plot()
```

```python
spectrum_coin = coin_wave.make_spectrum()
spectrum_coin.plot_power(linewidth=1,color = 'green')
thinkplot.config(xlabel='frequency (Hz)',
                 ylabel='power density',
                 xscale='log',
                 yscale='log',
                 ylim=[1e0, 1e11],
                 legend=False)
```

Looks like Brownian or red noise - over about 2 orders of magnitude for frequency, the power density decreases by 4 orders of magnitude.  Brownian noise seems to be the null hypothesis for financial data. [Copied from my ps02]
