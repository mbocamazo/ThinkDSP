2) In Chapter 7 I represented the DFT as matrix multiplication, but the more common presentation is as a summation:

From http://en.wikipedia.org/wiki/Discrete_Fourier_transform:

X_k\ \stackrel{\text{def}}{=}\ \sum_{n=0}^{N-1} x_n \cdot e^{-i 2 \pi k n / N}, \quad k\in\mathbb{Z}\,

Read the Wikipedia page on DFT.  Write an implementation of DFT based on this summation and compare your results to my implementation and/or numpy.fft.fft.

```python
#Allen's DFT:
def dft(ys):
    N = len(ys)
    ts = numpy.arange(N) / N
    freqs = numpy.arange(N)
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    amps = M.conj().transpose().dot(ys)
    return amps
```

```python
i = complex(0, 1)
PI2 = 2*math.pi
def own_dft(ys):
    """Implementation of dft based on summation"""
    N = len(ys)
    amps = numpy.ones((N, 1))
    amps = numpy.multiply(amps, i) # to cast as complex number
    for k in range(N): # for each sinusoidal component
        for n in range(N): # loop through the data points and find the agreement, summing over them
            amps[k] += ys[n]*numpy.exp(-i*PI2*k*n/float(N))
        amps[k] -= i # it works
    amps = numpy.reshape(amps, (1,N))
    return amps
        
```

```python
def ditfft2(x, N, s): 
    #x is the signal, N is len(x), and s is the stride of the array which should be 1... not accessing memory locations directly
    #N = len(x) #presumably
    amps = numpy.ones((N, 1))
    amps = numpy.multiply(amps, i) # to cast as complex number    
    if N == 1:
        return own_dft(x)
    else:
        amps[:N/2] = ditfft2(x, N/2, 2*s)
        amps[N/2:] = ditfft2(x+s, N/2, 2*s)
        for k in range(int(N/2-1)):
            t = amps[k]
            amps[k] = t + numpy.exp(-PI2*i*k/float(N))*amps[k+N/2]
            amps[k+N/2] = t - numpy.exp(-PI2*i*k/float(N))*amps[k+N/2]
    return amps       
        
```
