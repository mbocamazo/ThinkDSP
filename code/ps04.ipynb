2) In Chapter 7 I represented the DFT as matrix multiplication, but the more common presentation is as a summation:

From http://en.wikipedia.org/wiki/Discrete_Fourier_transform:

X_k\ \stackrel{\text{def}}{=}\ \sum_{n=0}^{N-1} x_n \cdot e^{-i 2 \pi k n / N}, \quad k\in\mathbb{Z}\,

Read the Wikipedia page on DFT.  Write an implementation of DFT based on this summation and compare your results to my implementation and/or numpy.fft.fft.

```python
#Allen's DFT:
def dft(ys):
    N = len(ys)
    ts = numpy.arange(N) / N
    freqs = numpy.arange(N)
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    amps = M.conj().transpose().dot(ys)
    return amps
```

```python
i = complex(0, 1)
PI2 = 2*math.pi
def own_dft(ys):
    """Implementation of dft based on summation"""
    N = len(ys)
    amps = numpy.ones((N, 1))
    amps = numpy.multiply(amps, i) # to cast as complex number
    for k in range(N): # for each sinusoidal component
        for n in range(N): # loop through the data points and find the agreement, summing over them
            amps[k] += ys[n]*numpy.exp(-i*PI2*k*n/float(N))
        amps[k] -= i # it works
    amps = numpy.reshape(amps, (1,N))
    return amps
        
```

4) Using the timing code from PS03, compare the performance of my implementation of autocorr from Chapter 5, my fft_autocorr from Chapter 8, and numpy.correlate.  Can you tell whether NumPy is using the FFT algorithm?

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```

```python
def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

def fft_convolve(signal, window):
    fft_signal = numpy.fft.fft(signal)
    fft_window = numpy.fft.fft(window)
    return numpy.fft.ifft(fft_signal * fft_window)

def fft_autocorr(signal):
    N = len(signal)
    window = signal[::-1]
    signal = zero_pad(signal, 2*N)
    window = zero_pad(window, 2*N)

    corrs = fft_convolve(signal, window)
    corrs = corrs[N//2: 3*N//2]
    return corrs
```

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)
ys = noise.ys
ns = 2 ** numpy.arange(4,14)
print(ns)
```

```python
lags, corrs = autocorr(noise)
thinkplot.plot(lags,corrs)
```

```python
results = []

for N in ns: #for each of the 2^n in range
    ys = noise.ys[:N]
    result = %timeit -o autocorr(thinkdsp.Wave(ys, framerate = 11025)) 
    results.append(result) #append time to compute 

bests_autocorr = [result.best for result in results]
bests_autocorr
```

```python
results = []

for N in ns: #for each of the 2^n in range
    ys = noise.ys[:N]
    result = %timeit -o fft_autocorr(ys) 
    results.append(result) #append time to compute 

bests_fft_autocorr = [result.best for result in results]
```

```python
results = []

for N in ns: #for each of the 2^n in range
    ys = noise.ys[:N]
    result = %timeit -o numpy.correlate(ys,ys) 
    results.append(result) #append time to compute 

bests_np_correlate = [result.best for result in results]
```

```python
thinkplot.plot(ns, bests_autocorr, label="autocorr")
thinkplot.plot(ns, bests_fft_autocorr, label="fft_autocorr")
thinkplot.plot(ns, bests_np_correlate, label="np.correlate")
thinkplot.config(xscale='log', yscale='log', loc = 2)
```

```python
import scipy.stats
x = numpy.log(ns)
y1 = numpy.log(bests_autocorr)
y2 = numpy.log(bests_fft_autocorr)
y3 = numpy.log(bests_np_correlate)
m1 = scipy.stats.linregress(x,y1)[0]
m2 = scipy.stats.linregress(x,y2)[0]
m3 = scipy.stats.linregress(x,y3)[0]
print(m1)
print(m2)
print(m3)
```
