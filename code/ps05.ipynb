4) Taking the difference of successive elements in a wave array has the same effect as convolution with the window [1, -1].  The filter for this
window is an increasing function of f, so you can think of diff as a high-pass filter.

One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f.

If we use a better approximation of a derivative, we might expect the corresponding filter to be a better approximation of a line.  The window [-1/2, 2, -3/2] is a second order forward difference approximation of the first derivative.  Compute its DFT and see what the filter looks like.

Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window?

See also http://en.wikipedia.org/wiki/Finite_difference_coefficient

```python
nrows = 1625
df = pandas.read_csv('coindesk-bpi-USD-close.csv', nrows=nrows, parse_dates=[0])
ys = df.Close.values
wave = thinkdsp.Wave(ys,11025)
```

```python
window = numpy.array([1.0, -1.0])
ys = scipy.signal.convolve(wave.ys, window, mode='valid')
diff = thinkdsp.Wave(ys, framerate=1)
diff.plot()
```

```python
window = numpy.array([-1/2, 2, -3/2])
ys = scipy.signal.convolve(wave.ys, window, mode='valid')
diff = thinkdsp.Wave(ys, framerate=1)
diff.plot()
```

looks about the same

```python
window = numpy.array([1.0, -1.0]) #simple deriv
window2 = numpy.array([-1, 4, -3]) / 2.0 #new deriv
padded = zero_pad(window, len(wave))
fft_window = thinkdsp.Wave(padded, framerate = 11025).make_spectrum()
#fft_window = numpy.fft.rfft(padded)
fft_window.plot(label = "[1 -1]")
#thinkplot.plot(fft_window, label = "[1 -1]")

padded2 = zero_pad(window2, len(wave))
fft_window2 = numpy.fft.rfft(padded2)
fft_window2 = thinkdsp.Wave(padded2, framerate = 11025).make_spectrum()
fft_window2.plot(label = "[-0.5 2 -1.5]")
#thinkplot.plot(abs(fft_window2), label = "[-0.5 2 -1.5]")

window3 = numpy.array([1, -2, 1])
padded3 = zero_pad(window3, len(wave))
fft_window3 = numpy.fft.rfft(padded3)
fft_window3 = thinkdsp.Wave(padded3, framerate = 11025).make_spectrum()
fft_window3.plot( label = "[1, -2, 1]")

thinkplot.config(xlabel = "Frequency, Hz", ylabel = "Amplitude", loc = 2)
```

The filter for the second derivative looks very similar to the second filter for the first derivative, but it is less extreme for lower frequencies, which should mean that it acts as a more agressive high-pass filter.  The diff filter is a less aggressive high-pass filter, judging from the ratio of the low and high frequencies.

6) Use spectrum.differentiate() to compute the derivative of a ParabolicSignal.  Hint: this works best if the duration of the wave is an integer number of periods.

```python
signal = thinkdsp.ParabolicSignal()
period = signal.period
wave = signal.make_wave(duration = period*10)
wave.plot()
```

```python
spectrum = wave.make_spectrum()
spectrum.plot()
```

```python
spectrum.differentiate()
spectrum.plot()
```
